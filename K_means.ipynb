```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def data_matrix(data):
    X = np.array(data)
    return X

def k_means(data,n_clusters):
    X = data_matrix(data)
    init_index = [np.random.choice([i for i in range(len(data))]) for _ in range(n_clusters)]
    init_centroides = np.array([X[i] for i in init_index])
    alloted_clusters_update = []
    for i in X:
        alloted_clusters_update.append(np.argmin([np.linalg.norm(i-j) for j in init_centroides]))
    alloted_clusters_update = np.array(alloted_clusters_update)
    alloted_clusters = alloted_clusters_update + 1
    k = 0
    while np.linalg.norm(alloted_clusters_update - alloted_clusters) > 1e-6:
        # init_centroides = init_centroides_update
        alloted_clusters_update = alloted_clusters
        # print(alloted_clusters_update)
        alloted_clusters = []
        for i in X:
            alloted_clusters.append(np.argmin([np.linalg.norm(i-j) for j in init_centroides]))
        alloted_clusters = np.array(alloted_clusters)
        # print(alloted_clusters)
        clusters = []
        for i in range(n_clusters):
            t = []
            for j in range(len(alloted_clusters)):
                if i == alloted_clusters[j]:
                    t.append(X[j])
            t = np.array(t)
            clusters.append(t)
        init_centroides = []
        for i in clusters:
            t = []
            for j in range(len(i[0])):
               t.append(np.average(i[:,j]))
            init_centroides.append(t)
        init_centroides = np.array(init_centroides)
        print("-------------------------------")
        print(f"Centroides at iteration {k+1}")
        print("-------------------------------")
        print(init_centroides)
        k = k+1
    if len(X[0]) == 2:
        for i in clusters:
            plt.scatter(i[:,0],i[:,1])
            plt.title("K-means Clustering Algorithm")
            plt.xlabel("Feature 1")
            plt.ylabel("Feature 2")
    elif len(X[0]) == 3:
        fig = plt.figure()
        ax = plt.axes(projection ='3d')
        for i in clusters:
            ax.scatter(i[:,0],i[:,1],i[:,2])
        plt.title("K-means Clustering Algorithm")
        
    else:
        print("Plotting is not possible!")
    return clusters
```

## Dataset 1


```python
data = pd.read_csv('data1.csv',header=None)
data = data.sample(frac=1)
```

```python
X = data_matrix(data)
```

```python
clusters = k_means(data,n_clusters=3)
```
